# Data Preprocessing

#### Finding Missing Values

```python
df.isnull().sum()
```

```python
fbs_frequency = df["fbs"].mode()[0]
fbs_frequency
```

```python
exang_frequency = df["exang"].mode()[0]
exang_frequency
```

```python
top_accuring_catagory=df["restecg"].value_counts()
top_accuring_catagory   
```

```python
slope_catagory=df["slope"].value_counts()
slope_catagory   
```

```python
thal_accuring_catagory=df["thal"].value_counts()
thal_accuring_catagory   
```

```python
df = df.fillna({
    "trestbps":df["trestbps"].mean(), 
    "chol":df["chol"].median(),
    "fbs": fbs_frequency,
    "thalch":df["thalch"].mean(),
    "exang": exang_frequency,
    "oldpeak":df["oldpeak"].mean(),
    "ca":df["ca"].median(),
    "num":df["num"].mean(),
    "restecg":df["restecg"].mode()[0],
    "slope":df["slope"].mode()[0],
    "thal":df["thal"].mode()[0]
},inplace=True)
```

```python
df.isnull().sum()
```

# EDA Analysis

#### Box plot

```python
fig = px.box(
    df, 
    x="num",
    y="trestbps",
    color="num",
    title="Distribution of Resting Blood Pressure by Heart Disease status",
    labels={
        "num": "Heart Disease",
        "tresrbps": "Resting Blood Pressure(mmHg)"
    }
) 
fig.show()
```

```python
fig = px.box(
    df, 
    x="exang",
    y="thalch",
    color="exang",
    title="Maximum Heart Rate Achieved by Exercise-Induced Angina",
    labels={
        "exang": "Exercise-Induced Angina",
        "thalch": "Maximum Heart Rate"
    }
) 
fig.show()
```

```python
fig = px.box(
    df, 
    x="num",
    y="chol",
    color="num",
    title="Distribution of Cholesterol Levels by Heart Disease status",
    labels={
        "num": "Heart Disease Status",
        "chol": "cholesterol(mg/DL)"
    }
) 
fig.show()
```

#### Histogram Graph

```python
df.hist(figsize=(15,12))
plt.show()
```

#### Heatmap Graph

```python
plt.figure (figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("correlation heatmap")
plt.show()
```

#### Pie Graph

```python
import plotly.express as px

# Replace numeric values with labels
df["Heart_Disease"] = df["num"].replace({
    0: "No Heart Disease",
    1: "Mild Heart Disease",
    2: "Moderate Heart Disease",
    3: "Severe Heart Disease",
    4: "Very Severe Heart Disease"
})

fig = px.pie(
    df,
    names="Heart_Disease",
    title="Heart Disease Status Distribution",
    hole=0.4
)
fig.update_layout(title_x=0.5,width=700,height=500)
fig.update_traces(textinfo="label+percent")

fig.show()
```

```python
import plotly.express as px

fig = px.pie(
    df,
    names="sex",
    title="Gender Distribution of Patients",
    hole=0.4
)
fig.update_layout(title_x=0.5,width=700,height=500)
fig.update_traces(textinfo="percent+label")

fig.show()
```
