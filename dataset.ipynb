# Data Preprocessing

#### Finding Missing Values

```python
df.isnull().sum()
```

```python
fbs_frequency = df["fbs"].mode()[0]
fbs_frequency
```

```python
exang_frequency = df["exang"].mode()[0]
exang_frequency
```

```python
top_accuring_catagory=df["restecg"].value_counts()
top_accuring_catagory   
```

```python
slope_catagory=df["slope"].value_counts()
slope_catagory   
```

```python
thal_accuring_catagory=df["thal"].value_counts()
thal_accuring_catagory   
```

```python
df = df.fillna({
    "trestbps":df["trestbps"].mean(), 
    "chol":df["chol"].median(),
    "fbs": fbs_frequency,
    "thalch":df["thalch"].mean(),
    "exang": exang_frequency,
    "oldpeak":df["oldpeak"].mean(),
    "ca":df["ca"].median(),
    "num":df["num"].mean(),
    "restecg":df["restecg"].mode()[0],
    "slope":df["slope"].mode()[0],
    "thal":df["thal"].mode()[0]
},inplace=True)
```

```python
df.isnull().sum()
```

#### Detecting Outliers

```python
import matplotlib. pyplot as plt 
import seaborn as sns
num_cols=df.select_dtypes (include='number').columns
for col in num_cols:
    plt.figure(figsize=(8,5))
    sns. boxplot(y=df[col])
    plt.title (f' box plot of {col}')
    plt.show()
```

# PIE GRAPH

```python
df["num"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(6,6)
)
plt.title("Heart Disease Distribution")
plt.ylabel("")
plt.show()
```

```python
fig=px.pie(df,names="restecg",title="pie graph",hole=0.4)
fig.update_traces(textinfo="percent+label")
fig.update_layout(title_x=0.5, template="plotly_white", width=700, height=500,legend_title="Education")
fig.show()
```
